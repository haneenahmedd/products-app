**Cells** are an independent 'unit'. When you click into a cell, you can 'run' it by clicking Shift + Enter, or by pressing the play (Run) button at the top of your notebook.

### Markdown Cells

Cells, can be plain text. In this book, for chapters where code is included, any place you see text - meaning places were we aren'twriting and executing code  - will be written in **Markdown**.

Markdown is a way to specify all the text formatting you see within the text itself.

#### _Italics_ & __Bold__

For example, *italicized text* can be specified with an \_underscore\_ or \*single asterisks\*. So, using markdown formatting, that text would appear as follows:

For example, italicized text can be specified with an _underscore_ or *single asterisks*.

**Bold text** requires \_\_two underccores\_\_ or \*\*two asterisks\*\* surrounding the text you would like to bold. Using Markdown formatting, that text would look like this:

**Bold text** requires __two underccores__ or **two asterisks** surrounding the text you would like to bold.



#### Headers

There are a number of different size header you can use with Markdown formatting. The number of pound signs preceeding the header specifies the size of the header.

For example, the following:

```
# Headers are specified with a pound sign
## The more pound signs, the smaller the header
#### But it's still larger
than just plain text.
```

would appear as follows using Markdown formatting:


## Headers are specified with a pound sign

### The more pound signs, the smaller the header

#### But it's still larger

than just plain text.

---

Note that to specify text for a header. the pound sign is followed by a space before the text for the header. The largest header -- specified by 1 pound sign -- is referred to as an H1 header. The second largest, H2, and so on and so forth.

#### Lists

Finally, **lists** are also possible with Markdown formatting. Ordered lists are specified with a number, followed by a decimal point, followed by a space and then the text for the list. For example, the following:

```
1. numbered item
2. item 2
3. item 3
```

would appear as follows with Markdown formatting:

1. numbered item
2. item 2
3. item 3

Bulleted or unordered lists are also possible. these are specified by either a dash (`-`) or an asterisk (`*`) instead of the number and decimal point.

For example, either of the following:

```
- item 1
- item 2
- item 3
```

or

```
* item 1
* item 2
* item 3
```

would appear as follows with Markdown formatting:

- item 1
- item 2
- item 3

#### Cell Order

Ok, so we've discussed that the numbers in the square brackets to the left of a cell show which cells have been run, and in what order and that an asterisk (`*`) means that the cell is currently running. What we haven't yet discussed is that the order in which the cells are run does ***not*** matter to Python or Jupyter notebooks.

This allows you to flexibly test and develop code. For example, say you've run a few cells of code in order from top to bottom. Then, you realize you wanted a different value in that first code cell. You can go back to that first cell, change it to be the value you want, and then return back to a different cell. Python and Jupyter Notebooks will keep track of whatever was run most recently.

For beginners and individuals less familiar with working in notebooks, this can sometimes take a bit to get used to, but the benefits of this flexibility outweigh the cognitive load it takes to remember that the order isn't set in stone to go from top to bottom.

````python
import os, glob, yaml

def list_structure(base, max_depth=3):
    for root, dirs, files in os.walk(base):
        depth = root.replace(base, '').count(os.sep)
        if depth > max_depth:
            continue
        indent = '  ' * depth
        print(f"{indent}{os.path.basename(root)}/")
        if depth == max_depth:
            for f in files[:5]:
                print(f"{'  ' * (depth+1)}{f}")

print("=" * 40)
print("Dataset 1 Structure:")
print("=" * 40)
list_structure('/content/ds1')

print("\n" + "=" * 40)
print("Dataset 2 Structure:")
print("=" * 40)
list_structure('/content/ds2')

print("\n" + "=" * 40)
print("YAML Class Info:")
print("=" * 40)
for yaml_file in glob.glob('/content/ds1/**/*.yaml', recursive=True) + \
                 glob.glob('/content/ds2/**/*.yaml', recursive=True):
    with open(yaml_file) as f:
        data = yaml.safe_load(f)
    print(f"\n📄 {yaml_file}")
    print(f"   names : {data.get('names')}")
    print(f"   nc    : {data.get('nc')}")
````

````python
import os, shutil, random
from pathlib import Path
import torch
from PIL import Image

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {DEVICE}")

# ──────────────────────────────────────────────────
# REMAP
# ──────────────────────────────────────────────────
DS1_REMAP = {
    0: 0,   # hardhat    → helmet (0)
    2: 1,   # no-hardhat → head   (1)
}

DS2_REMAP = {
    0: 0,   # helmet → helmet (0)
    1: 1,   # head   → head   (1)
}

OUTPUT_DIR = '/content/helmet_merged'
SEED       = 42
IMG_EXTS   = {'.jpg', '.jpeg', '.png', '.bmp'}

# DS1 split ratios → train / val / test
DS1_SPLIT  = (0.70, 0.20, 0.10)

# DS2 split ratios → train / test (no val)
DS2_SPLIT  = (0.80, 0.20)

# ──────────────────────────────────────────────────
# Collect DS1 — standard images/labels mirror
# ──────────────────────────────────────────────────
def collect_pairs_ds1(base_dir):
    pairs = []
    base  = Path(base_dir)
    for img_path in base.rglob('*'):
        if img_path.suffix.lower() not in IMG_EXTS:
            continue
        lbl_path = Path(
            str(img_path).replace('images', 'labels')
        ).with_suffix('.txt')
        if lbl_path.exists():
            pairs.append((img_path, lbl_path))
    return pairs

# ──────────────────────────────────────────────────
# Collect DS2 — custom structure:
#   images/train/ → labels/train_labels/
#   images/val/   → labels/val_labels/
# ──────────────────────────────────────────────────
def collect_pairs_ds2(base_dir):
    pairs     = []
    base      = Path(base_dir)
    split_map = {
        'train' : 'train_labels',
        'val'   : 'val_labels',
    }
    for split, lbl_folder in split_map.items():
        img_dir = base / 'helmet_dataset_new' / 'images' / split
        lbl_dir = base / 'helmet_dataset_new' / 'labels' / lbl_folder

        if not img_dir.exists():
            print(f"  ⚠️  DS2 image dir not found: {img_dir}")
            continue
        if not lbl_dir.exists():
            print(f"  ⚠️  DS2 label dir not found: {lbl_dir}")
            continue

        for img_path in img_dir.iterdir():
            if img_path.suffix.lower() not in IMG_EXTS:
                continue
            lbl_path = lbl_dir / (img_path.stem + '.txt')
            if lbl_path.exists():
                pairs.append((img_path, lbl_path))

    return pairs

# ──────────────────────────────────────────────────
def remap_label(src, dst, remap):
    lines_out = []
    with open(src) as f:
        for line in f:
            parts = line.strip().split()
            if not parts:
                continue
            cls_id = int(parts[0])
            if cls_id in remap:
                lines_out.append(f"{remap[cls_id]} {' '.join(parts[1:])}\n")
    if lines_out:
        dst.parent.mkdir(parents=True, exist_ok=True)
        with open(dst, 'w') as f:
            f.writelines(lines_out)
        return True
    return False

def validate_image_gpu(img_path):
    try:
        import torchvision.transforms.functional as TF
        img    = Image.open(img_path).convert('RGB')
        tensor = TF.to_tensor(img).to(DEVICE)
        return tensor.shape[0] == 3
    except Exception:
        return False

# ──────────────────────────────────────────────────
# Collect all pairs
# ──────────────────────────────────────────────────
pairs_ds1 = collect_pairs_ds1('/content/ds1')
pairs_ds2 = collect_pairs_ds2('/content/ds2')

print(f"Dataset 1 total pairs : {len(pairs_ds1)}")
print(f"Dataset 2 total pairs : {len(pairs_ds2)}")

# ──────────────────────────────────────────────────
# Split DS1 → train / val / test  (70/20/10)
# ──────────────────────────────────────────────────
random.seed(SEED)
random.shuffle(pairs_ds1)

n1_train = int(len(pairs_ds1) * DS1_SPLIT[0])
n1_val   = int(len(pairs_ds1) * DS1_SPLIT[1])

ds1_train = pairs_ds1[:n1_train]
ds1_val   = pairs_ds1[n1_train : n1_train + n1_val]
ds1_test  = pairs_ds1[n1_train + n1_val:]

print(f"\nDS1 → Train : {len(ds1_train)} | Val : {len(ds1_val)} | Test : {len(ds1_test)}")

# ──────────────────────────────────────────────────
# Split DS2 → train / test  (80/20) — no val
# ──────────────────────────────────────────────────
random.shuffle(pairs_ds2)

n2_train = int(len(pairs_ds2) * DS2_SPLIT[0])

ds2_train = pairs_ds2[:n2_train]
ds2_test  = pairs_ds2[n2_train:]

print(f"DS2 → Train : {len(ds2_train)} | Test : {len(ds2_test)}")

# ──────────────────────────────────────────────────
# Merge splits
# train = DS1 train + DS2 train
# val   = DS1 val only
# test  = DS1 test + DS2 test
# ──────────────────────────────────────────────────
splits = {
    'train' : [(p, DS1_REMAP) for p in ds1_train] +
              [(p, DS2_REMAP) for p in ds2_train],

    'val'   : [(p, DS1_REMAP) for p in ds1_val],

    'test'  : [(p, DS1_REMAP) for p in ds1_test] +
              [(p, DS2_REMAP) for p in ds2_test],
}

print(f"\nFinal merged splits:")
print(f"  Train : {len(splits['train'])}  (DS1 train + DS2 train)")
print(f"  Val   : {len(splits['val'])}   (DS1 val only)")
print(f"  Test  : {len(splits['test'])}   (DS1 test  + DS2 test)")

# ──────────────────────────────────────────────────
# Clean old output dir
# ──────────────────────────────────────────────────
if Path(OUTPUT_DIR).exists():
    shutil.rmtree(OUTPUT_DIR)
    print(f"\n🗑️  Cleared old output dir: {OUTPUT_DIR}")

# ──────────────────────────────────────────────────
# Copy files into output structure
# ──────────────────────────────────────────────────
for split, items in splits.items():
    img_dir = Path(OUTPUT_DIR) / 'images' / split
    lbl_dir = Path(OUTPUT_DIR) / 'labels' / split
    img_dir.mkdir(parents=True, exist_ok=True)
    lbl_dir.mkdir(parents=True, exist_ok=True)

    copied, skipped, invalid = 0, 0, 0
    for idx, ((img_path, lbl_path), remap) in enumerate(items):
        stem    = f"{img_path.stem}_{idx}"
        dst_img = img_dir / f"{stem}{img_path.suffix}"
        dst_lbl = lbl_dir / f"{stem}.txt"

        if not validate_image_gpu(img_path):
            invalid += 1
            continue

        if remap_label(lbl_path, dst_lbl, remap):
            shutil.copy2(img_path, dst_img)
            copied += 1
        else:
            skipped += 1

    print(f"  {split:6s} → {copied} copied | {skipped} skipped | {invalid} invalid")

print("\n✅ Merge & split complete!")

# ──────────────────────────────────────────────────
# Final summary
# ──────────────────────────────────────────────────
print("\n" + "=" * 50)
print("            FINAL DATASET SUMMARY")
print("=" * 50)
print(f"  {'Split':<8} {'Images':>8}   {'Source'}")
print(f"  {'-'*40}")
for split in ['train', 'val', 'test']:
    img_count = len(list((Path(OUTPUT_DIR) / 'images' / split).iterdir()))
    source = {
        'train' : 'DS1 (70%) + DS2 (80%)',
        'val'   : 'DS1 (20%) only',
        'test'  : 'DS1 (10%) + DS2 (20%)',
    }[split]
    print(f"  {split:<8} {img_count:>8}   {source}")
print("=" * 50)
````

````python
import yaml

OUTPUT_DIR = '/content/helmet_merged'

config = {
    'path'  : OUTPUT_DIR,
    'train' : 'images/train',
    'val'   : 'images/val',
    'test'  : 'images/test',
    'nc'    : 2,
    'names' : ['helmet', 'head']
}

yaml_path = f'{OUTPUT_DIR}/data.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(config, f, default_flow_style=False)

print(f"✅ data.yaml saved → {yaml_path}")
with open(yaml_path) as f:
    print(f.read())
````

````python
from ultralytics import YOLO
import torch

print(f"Training on: {torch.cuda.get_device_name(0)}")

model = YOLO('yolo11n.pt').to('cuda')

model.train(
    data      = yaml_path,
    epochs    = 40,
    imgsz     = 640,
    batch     = 32,
    device    = 1,        # T4 GPU
    workers   = 4,
    cache     = False,     # cache images in RAM
    amp       = True,     # mixed precision — faster on T4
    name      = 'helmet_detector',
    project   = '/content/runs',
    exist_ok  = True,
)
````

````python
import torch

print(f"Validating on: {torch.cuda.get_device_name(0)}")

metrics = model.val(
    data   = yaml_path,
    device = 1,
    batch  = 32,
    imgsz  = 640,
)

print(f"\n{'='*40}")
print(f"mAP50        : {metrics.box.map50:.4f}")
print(f"mAP50-95     : {metrics.box.map:.4f}")
print(f"Precision    : {metrics.box.p.mean():.4f}")
print(f"Recall       : {metrics.box.r.mean():.4f}")
print(f"{'='*40}")
````

````python
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from ultralytics import YOLO
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)

DEVICE     = 0  # T4 GPU
CONF_THRES = 0.5
IOU_THRES  = 0.5
CLASS_NAMES = ['helmet', 'head']

model = YOLO('/content/runs/helmet_detector/weights/best.pt')
model.to('cuda')

test_img_dir = Path(f'{OUTPUT_DIR}/images/test')
test_lbl_dir = Path(f'{OUTPUT_DIR}/labels/test')

IMG_EXTS = {'.jpg', '.jpeg', '.png', '.bmp'}


y_true = []
y_pred = []

image_files = sorted([
    f for f in test_img_dir.iterdir()
    if f.suffix.lower() in IMG_EXTS
])

print(f"Running inference on {len(image_files)} test images...")

for img_path in image_files:
    lbl_path = test_lbl_dir / (img_path.stem + '.txt')
    if not lbl_path.exists():
        continue

    # ── Ground truth ──
    with open(lbl_path) as f:
        gt_classes = [int(line.split()[0]) for line in f if line.strip()]

    # ── Predictions ──
    results = model.predict(
        source  = str(img_path),
        device  = DEVICE,
        conf    = CONF_THRES,
        iou     = IOU_THRES,
        imgsz   = 640,
        verbose = False,
    )

    pred_classes = []
    for r in results:
        if r.boxes is not None and len(r.boxes):
            pred_classes = r.boxes.cls.cpu().numpy().astype(int).tolist()

    # ── Match GT ↔ Pred by count per class ──
    # For each GT class, check if prediction contains it
    for gt_cls in gt_classes:
        y_true.append(gt_cls)
        if gt_cls in pred_classes:
            y_pred.append(gt_cls)      # correct detection
            pred_classes.remove(gt_cls)
        else:
            # missed detection → predict most common pred or -1
            y_pred.append(
                pred_classes.pop(0) if pred_classes else -1
            )

    # ── Extra false positives ──
    for fp_cls in pred_classes:
        y_true.append(-1)
        y_pred.append(fp_cls)

print(f"Total GT instances   : {len(y_true)}")
print(f"Total Pred instances : {len(y_pred)}")
````

````python
# Filter out -1 (background/unmatched) for clean report
mask   = [i for i, v in enumerate(y_true) if v != -1]
y_true_f = [y_true[i] for i in mask]
y_pred_f = [y_pred[i] for i in mask]

# Clip pred to valid class range
y_pred_f = [
    c if c in range(len(CLASS_NAMES)) else 0
    for c in y_pred_f
]

print("=" * 55)
print("          CLASSIFICATION REPORT")
print("=" * 55)
report = classification_report(
    y_true_f,
    y_pred_f,
    target_names=CLASS_NAMES,
    digits=4,
    zero_division=0
)
print(report)
````

````python
cm = confusion_matrix(y_true_f, y_pred_f, labels=list(range(len(CLASS_NAMES))))

# ── Raw counts ──
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=CLASS_NAMES,
    yticklabels=CLASS_NAMES,
    ax=axes[0],
    linewidths=0.5,
    linecolor='gray'
)
axes[0].set_title('Confusion Matrix (Counts)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Predicted Label', fontsize=12)
axes[0].set_ylabel('True Label', fontsize=12)

# ── Normalized (%) ──
cm_norm = cm.astype('float') / cm.sum(axis=1, keepdims=True)
cm_norm = np.nan_to_num(cm_norm)

sns.heatmap(
    cm_norm,
    annot=True,
    fmt='.2%',
    cmap='Greens',
    xticklabels=CLASS_NAMES,
    yticklabels=CLASS_NAMES,
    ax=axes[1],
    linewidths=0.5,
    linecolor='gray'
)
axes[1].set_title('Confusion Matrix (Normalized)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Predicted Label', fontsize=12)
axes[1].set_ylabel('True Label', fontsize=12)

plt.suptitle('Helmet Detection — YOLOv11 Test Results', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()

# Save
save_path = '/content/runs/confusion_matrix.png'
plt.savefig(save_path, dpi=150, bbox_inches='tight')
plt.show()
print(f" Confusion matrix saved → {save_path}")
````
